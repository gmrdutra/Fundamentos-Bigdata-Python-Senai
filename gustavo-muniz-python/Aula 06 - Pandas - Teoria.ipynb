### Renomear as colunas

```python
# Todas as colunas
# Modificar os nomes
# As colunas devem estar na mesma ordem dos nome
# A quantidade de nomes deve ser igual a quantidade de colunas

df.columns = ['idade','trabalho','estado_civil','educacao','default','balanco','housing','empréstimo','contato','dia','mês','duração','campanha','pdays','previous','poutcome']

df.head()
```

###  Renomear coluna específica

```python
df = df.rename(columns = { 
    'idade' : 'IDADE',
    'trabalho'   : 'Profissione'
})

df.head()
```
