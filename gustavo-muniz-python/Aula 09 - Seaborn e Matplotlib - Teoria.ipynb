### Primeiros passos com Matplotlib 

```python
# Importando as bibliotecas 

import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 

# Criando os arrays (matrizes) para criar os gráficos 

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um gráfico com o valores anteriores 

plt.plot(x,y)
```

### MODIFICANDO O TIPO DE MARCADOR

```python
plt.plot(x,y,marker ='X', c = 'red')
```

### Modificando a forma como os pontos são ligados e o tipo de linha

```python
plt.plot(
    x,
    y,
    marker = '*',
    c = 'blue',
    ds = 'steps', # Forma de ligação
    ls = '--' # Estilo da linha
)
```

### Alterando as cores dos marcadores 

```python
plt.plot(
    x, # Vetor responsavel pelos valores do eixo X
    y, # Vetor responsavel pelos valores do eixo y
    marker = '*', #Marcador do gráfico
    c = 'blue', # Cor da linha
    ds = 'steps', # Forma de ligação
    ls = '--', # Estilo da linha
    mec = 'red' # Cor dos marcadores
)
```

### Adicionando outros elementos ao gráfico 

```python
plt.plot(x,y) # Criando o gráfico 

# Adicionando o título do Eixo X

plt.xlabel('Valores de X', fontsize = 12, c = 'green')

# Adicionando o título do Eixo Y

plt.ylabel('Valores de Y', fontsize = 12, c ='red')

# Adicionando o título do Gráfico

plt.title('Título do Gráfico', fontsize = 16, c = 'purple')

plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'brown')
```

```python
# Gerar dados
x = np.linspace (0,2 * np.pi, 100) # Gerar 100 numeros entre 0 e 2 * PI
y = np.sin(x) # Função Seno
z = np.cos(x) # Função Cosena

# Curvas do Gráfico 

plt.plot (x,y,label = "Função Seno", color = '#853DF2')
plt.plot (x,z,label = "Função Coseno", color = '#F53EE4')

# Adicionar uma Legenda ao Gráfico

plt.legend()

# Adicionar um titulo ao Gráfico
plt.title ('Função Trigonométricas')

# Adicionar um titulo ao Eixo

plt.xlabel ('Valores de X')
plt.ylabel ('Valores de Y')

# Salvar o Gráfico como Imagem

plt.savefig('./Gráficos/funcoes-trigonometrixa.png')
```

### Multiplos Subgráficos 


```python
# Criando dois gráficos na Horizontal (linha)

# plt.subplot ( nrows = Número de linhas , ncols = numero de colunas, pos = posição do gráfico  )

# Primeiros Gráfico
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Subplot 01')

# Segundo Gráfico
plt.subplot (2,1,2)
plt.plot (x,z)
plt.title('Subplot 02')
```

### Dois Gráficos na vertical (Colunas)

```python
# Primeiro Gráfico 
plt.subplot (1,2,1)
plt.plot (x,y)

# Segundo Gráfico 
plt.subplot(1,2,2)
plt.plot (x,z)
```

### Exemplo - 01 - Gráfico de Pizza

```python
# Importar as Bibliotecas

import matplotlib.pyplot as plt

# Dados para plotar no Gráfico 

languages = ['Java', 'Python' , 'PHP', 'JavaScript', 'C#', 'C++']
popularidade = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
exp = [0.1,0,0,0,0,0]
cores = ['orange','purple', 'white', 'green' , 'yellow', 'gray']

# Gerar o Gráfico
plt.pie(popularidade, labels = languages, autopct = '%1.1f%%', shadow = True, explode = exp, colors = cores)

```

```python
x = np.array([-3,-2.5,-2,-1,0,1,2,2.5,3])
y = np.array([-27,-15.62,-8,-1,0,1,8,15.62,27])

plt.plot (x,y, c = 'magenta' , label = 'Função Cúbica', marker = 'o', ls = '--', lw = 1)

plt.xlabel('Valor de X', fontsize = 15)
plt.ylabel('Valor de Y',fontsize = 15.5)

plt.legend (fontsize = 20, loc = 'lower right')

```
