```python
# Filtrando dados de uma forma mais fácil 

# 1 - Selecionar a coluna de interesse

df_cargos = df['Cargo'].drop_duplicates()

# 2 - Converter o df_cargos em lista

cargos = df_cargos.to_list()
cargos
```

```python
for cargo in cargos:
    df_final = df[df['Cargo'] == cargo]
    caminho = '../gustavo-muniz-python/Relatório - Exercicio 3/'
    arquivo = cargo + '.xlsx'
    df_final.to_excel(caminho + arquivo)
```

```python
df_vendedores = df['Vendedor'].drop_duplicates()

vendedores = df_vendedores.to_list()

for vendedor in vendedores :
    df_final = df[df['Vendedor'] == vendedor]
    caminho = '../gustavo-muniz-python/Relatório - Exercicio 3/Vendedor/'
    arquivo = vendedor + '.xlsx'
    df_final.to_excel (caminho + arquivo)
```

```python
df_clientes = df['Cliente'].drop_duplicates ()

lista_clientes = df_clientes.to_list()

for cliente in lista_clientes :
    df_final = df[df['Cliente'] == cliente]
    caminho = '../gustavo-muniz-python/Relatório - Exercicio 3/Cliente/'
    arquivo = cliente + ".xlsx"
    df_final.to_excel ( caminho + arquivo)
```

```python
df_UF = df['UF'].drop_duplicates()

UFs = df_UF.to_list()

for estados in UFs :
    df_final = df[df['UF'] == estados]
    caminho = '../gustavo-muniz-python/Relatório - Exercicio 3/UF/'
    arquivo = estados + ".xlsx"
    df_final.to_excel (caminho + arquivo)
```
