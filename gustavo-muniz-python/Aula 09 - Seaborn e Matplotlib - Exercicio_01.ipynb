```python
# Importar as bibliotecas

import pandas as pd 
import matplotlib.pyplot as plt
import os 

# Mapear o caminho do arquivo 

caminho = './Arquivos/company_sales_data.csv'

# Alocar os dados em uma DataFrame

df = pd.read_csv ( caminho )

# Visualizar o cabeçalho do DataFrame

df.head ()

#Selecionar os dados que vão ser apresentados no gráfico

listaLucro = df['total_profit'].to_list()
listaMeses = df['month_number'].to_list()

# Criar o gráfico utilizando os dados listados anteriormente

plt.plot(listaMeses,listaLucro, label = 'Lucro Mensal do Ano Anterior')

plt.xlabel('Número do Mês')
plt.ylabel('Lucro')
plt.title('Lucro Mensal da Empresa')
plt.legend() 


```
