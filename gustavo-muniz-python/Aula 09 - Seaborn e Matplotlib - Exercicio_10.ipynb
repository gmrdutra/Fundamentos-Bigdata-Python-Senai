```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns

caminho = './Arquivos/DadosVendaCarros.xlsx'

df = pd.read_excel (caminho)

df.columns = ['DataNotaFiscal','Fabricante','Estado','ValorVenda','ValorCusto','TotalDesconto','CustoEntrega','CustoMaoDeObra','NomeCliente','Modelo','Cor','Ano','Mês']

df = df.dropna()

df_ano = df.groupby('Ano')['ValorVenda'].sum()

anos = df_ano.index
vendas = df_ano.values

plt.bar(anos,vendas)

```

```python
#Filtrando as vendas de 2016
df_2016 = df[df['Ano'] == 2016]

# Agrupar as vendas por mês
df_mes = df_2016.groupby('Mês')['ValorVenda'].sum()

mes = df_mes.index
vendas_2016 = df_mes.values

plt.bar(mes,vendas_2016)
```
